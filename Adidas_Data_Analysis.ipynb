```python
top_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Adidas Product Categories")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.show()
```

```python
category_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=category_rating.values, y=category_rating.index)
plt.title("Average Rating by Product Category")
plt.show()
```

```python
## Key Insights

After analyzing the Adidas global product dataset, several interesting patterns were observed.

First, Adidas offers a wide range of products across different categories such as lifestyle, sportswear, and performance gear. A large portion of the products falls into the lifestyle and sports categories, showing that Adidas focuses heavily on everyday athletic and casual wear.

Second, most products have very high customer ratings, with the average rating being above 4.5. This suggests that customers are generally very satisfied with Adidas products and the brand maintains strong product quality.

Another interesting observation is related to pricing. The dataset shows a wide price range, from very low-priced items to premium products. However, most items fall within a moderate price range, making them accessible to a large group of customers.

Discount patterns also reveal that while many products are sold at regular price, some products receive significant discounts, sometimes reaching up to 60%. These discounts likely help boost sales during promotional periods.

Finally, the dataset shows that Adidas products are available across multiple international markets. Some products appear in several markets, indicating global demand and wide distribution of certain popular items.

Overall, the analysis highlights that Adidas maintains a strong presence in the sportswear market by offering highly rated products, competitive pricing, and global availability.
```
